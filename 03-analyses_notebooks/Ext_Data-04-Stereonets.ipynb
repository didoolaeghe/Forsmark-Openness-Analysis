# Orientations stereonets

We present functions used to plot fracture orientations using stereonets. First we show how to make a stereonet of all fracture and open fractures. Then we show how to make a stereonet representing the open fraction, or openness.

```python
import sys, os, math, numpy as np
import pandas as pd
import matplotlib.pyplot as plt


path=r'python_functions'
sys.path.append(path)
import utils_stereonet as uS
```

```python
input_folder = r'input_data'
filename = 'p_frac_core_select'
all_data = pd.read_csv(os.path.join(input_folder,filename+str('.txt')), sep = ';')
```

### 1) Stereonet of all and open data

```python
#all data
dip_all = all_data['DIP(degrees)'].values
strike_all = all_data['STRIKE(degrees)'].values
alpha_all = all_data['ALPHA(degrees)'].values
fig = uS.plot_stereonet1(dip_all, strike_all, alpha = alpha_all, show_poles = False) #try plot_stereonet2 for home-made stereonet function
fig.suptitle('All data', fontsize = 20)
fig.set_size_inches(400./fig.dpi,400./fig.dpi)
plt.show()

#open data
open_data = all_data.loc[all_data['FRACT_INTERPRET'].isin(['Open','Partly open'])]
dip_open = open_data['DIP(degrees)'].values
strike_open = open_data['STRIKE(degrees)'].values
alpha_open = open_data['ALPHA(degrees)'].values
fig = uS.plot_stereonet1(dip_open, strike_open, alpha = alpha_open, show_poles = False)
fig.suptitle('Open data', fontsize = 20)
fig.set_size_inches(400./fig.dpi,400./fig.dpi)
plt.show()
```

### 2) Open fraction stereonet

```python
fig = uS.plot_stereonet_fraction(dip_open, strike_open, dip_all, strike_all, alpha1 = alpha_open, alpha2 = alpha_all, cbmax = 0.7)
fig.set_size_inches(400./fig.dpi,400./fig.dpi)
plt.show()
```
